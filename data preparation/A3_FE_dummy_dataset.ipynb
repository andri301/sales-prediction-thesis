## Extend the dataframe to accommodate lags

```python
data[['Year', 'Quarter']] = data['quarter'].str.split('-', expand=True)
data['Year'] = data['Year'].astype(int)
data['Quarter'] = data['Quarter'].str.extract('(\d+)').astype(int)
data = data.drop('quarter', axis=1)
# Rearrange columns
data = data[['Year', 'Quarter'] + [col for col in data.columns if col not in ['Year', 'Quarter']]]
data
```

```python
# Determine the last year and quarter in the DataFrame
last_year = data['Year'].iloc[-1]
last_quarter = data['Quarter'].iloc[-1]

# Calculate the next quarter
if last_quarter == 4:
    next_year = last_year + 1
    next_quarter = 1
else:
    next_year = last_year
    next_quarter = last_quarter + 1

# Extend the DataFrame for the next four quarters
for i in range(4):
    new_row = {'Year': next_year, 'Quarter': next_quarter}
    data = pd.concat([data, pd.DataFrame([new_row])], ignore_index=True)
    if next_quarter == 4:
        next_year += 1
        next_quarter = 1
    else:
        next_quarter += 1
```

## Correlation analysis for best lags

```python
lagged_columns = data.columns.tolist()[3:]

# Initialize an empty list to store lagged attribute groups
lagged_attribute_groups = []

# Iterate over the lagged attribute column names
for col in lagged_columns:
    # Extract the common prefix up to the underscore before the lag number
    prefix = col.split('_lag_')[0] + '_lag_'
    # Add the prefix to the lagged attribute groups list if not already present
    if prefix not in lagged_attribute_groups:
        lagged_attribute_groups.append(prefix)

# Now, lagged_attribute_groups contains the desired groups of lagged attributes
print(lagged_attribute_groups)
```

```python
# Define a function to find the best correlated lag for each attribute group
def find_best_correlated_lag(df, target_variable, lagged_attributes):
    best_correlations = {}
    for attr in lagged_attributes:
        lag_correlations = {}
        for lag_col in df.columns:
            if lag_col.startswith(attr):
                lag_correlation = df[target_variable].corr(df[lag_col])
                lag_correlations[lag_col] = lag_correlation
        best_lag = max(lag_correlations, key=lag_correlations.get)
        best_correlations[attr] = (best_lag, lag_correlations[best_lag])
    return best_correlations

# Assuming your target variable is 'NCN_sales_est'
target_variable = 'NCN_sales_est'

# Finding best correlated lag for each attribute group
best_correlated_lags = find_best_correlated_lag(data, target_variable, lagged_attribute_groups)

# Displaying the results
for attr, (best_lag, correlation) in best_correlated_lags.items():
    print(f"Best correlated lag for attribute group {attr}: {best_lag}, Correlation: {correlation}")
```

```python
# Initialize empty lists to store data
attributes = []
best_lag_values = []
weighted_lags = []

# Extract attribute names, best lag values, and calculate weighted lags
for attribute, (lag_name, correlation) in best_correlated_lags.items():
    # Remove '_lag_' from attribute name
    attribute_name = attribute.replace('_lag_', '')
    attributes.append(attribute_name)
    # Extract the lag value as an integer from lag_name
    lag_value = int(lag_name.split('_')[-1])
    best_lag_values.append(lag_value)
    # Calculate absolute value of weighted lag
    weighted_lag = abs(lag_value * correlation)
    weighted_lags.append(weighted_lag)

# Create DataFrame
df = pd.DataFrame({'Attribute': attributes, 'Best_Lag_Value': best_lag_values, 'Weighted_lag': weighted_lags})
df
```

```python
# Calculate the mean of 'Best_Lag_Value' column
mean_lag_value = df['Best_Lag_Value'].mean()
mean_wei_value = df['Weighted_lag'].mean()
print("Mean of Best_Lag_Value:", mean_lag_value)
print("Mean of Weighted_Lag:", mean_wei_value)
```

## Create engineered values

Engineered features are created for the values of the best correlated lag. They will capture:
1. Characteristics of time series:
- moving average of last 4 quarters
- moving standard deviation of last for quarters
2. Log transform of the values
3. Growth indicators:
- absolute, ie the difference between consecutive values
- relative. ie the ratio between consecutive values

```python
# For log transform, check attributes with 0 or negative values
# Use describe() method to get summary statistics
summary_stats = data.describe()

# Create a boolean mask to identify columns with 0 or negative values
zero_or_negative_mask = (summary_stats.loc['min'] <= 0)

# Extract the column names that meet the condition
columns_with_zero_or_negative_values = zero_or_negative_mask[zero_or_negative_mask].index.tolist()

# Display the column names with 0 or negative values
print("Columns with 0 or negative values:")
print(columns_with_zero_or_negative_values)
```

```python
# Iterate over the best correlated lags
for group, (best_correlated_lag, correlation) in best_correlated_lags.items():
    # Apply feature engineering operations for the best correlated lag
    # Moving average of last 4 quarters
    data[f'Avg4_{best_correlated_lag}'] = (data[best_correlated_lag] + data[best_correlated_lag].shift() + data[best_correlated_lag].shift(2) + data[best_correlated_lag].shift(3)) / 4

    # Moving standard deviation of last 4 quarters
    data[f'Std4_{best_correlated_lag}'] = data[best_correlated_lag].rolling(window=4).std()

    # Log transform of the values (conditionally)
    if best_correlated_lag not in columns_with_zero_or_negative_values:
        data[f'Log_{best_correlated_lag}'] = np.log(data[best_correlated_lag])

    # Absolute difference between consecutive values
    data[f'Diff_{best_correlated_lag}'] = data[best_correlated_lag].diff()

    # Relative ratio between consecutive values
    data[f'Div_{best_correlated_lag}'] = data[best_correlated_lag].div(data[best_correlated_lag].shift())

# Display the DataFrame with the new features
data.head(15)
```
